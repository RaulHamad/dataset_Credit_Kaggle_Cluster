### 🐼 Pandas (`pd`)
O **Pandas** é a biblioteca padrão para manipulação e análise de dados.
* **Função:** Permite trabalhar com estruturas de dados de alto desempenho, como os *DataFrames*.
* **Uso no projeto:** Limpeza de dados, filtragem, agregação e leitura de arquivos (CSV/Excel).

### 📈 Matplotlib (`plt`)
O **Matplotlib** é a biblioteca base para a geração de gráficos em Python.
* **Função:** Oferece controle total sobre todos os elementos de uma figura (eixos, legendas, cores).
* **Uso no projeto:** Construção da estrutura dos gráficos e personalizações detalhadas de layout.

### 🎨 Seaborn (`sns`)
Construída sobre o Matplotlib, o **Seaborn** é focado em visualização de dados estatísticos.
* **Função:** Facilita a criação de gráficos complexos e esteticamente atraentes com menos código.
* **Uso no projeto:** Criação de mapas de calor (heatmaps), distribuições e gráficos de correlação.

### 📊 Visualização da Distribuição de Clusters

Nesta etapa, geramos um gráfico de barras (Countplot) para visualizar como os clientes foram agrupados pelo algoritmo K-Means e qual a proporção de cada grupo.

````python

# 2. Configurar o estilo visual
sns.set_theme(style="whitegrid")
plt.figure(figsize=(10, 6))

# 3. Criar o gráfico de contagem (Countplot)
ax = sns.countplot(data=df_k2, x='cluster_kmeans', palette='viridis')

# 4. Adicionar detalhes informativos
plt.title('Distribuição das Instâncias por Cluster (kmeans 2)', fontsize=15)
plt.xlabel('Cluster', fontsize=12)
plt.ylabel('Quantidade de Clientes', fontsize=12)

# Adicionar os números exatos em cima das barras
for p in ax.patches:
    ax.annotate(f'{int(p.get_height())}',
                (p.get_x() + p.get_width() / 2., p.get_height()),
                ha = 'center', va = 'center',
                xytext = (0, 9),
                textcoords = 'offset points')

plt.show()

# 5. Exibir a distribuição percentual no console
print("Distribuição das instâncias:")
print(df_k2['cluster_kmeans'].value_counts(normalize=True) * 100)
````

### 🏷️ Tradução e Renomeação das Colunas

#### Código para Renomeação:

### 📈 Análise de Tendência Central e Dispersão (K-Means K=2)

Para extrair o perfil detalhado de cada cluster, calculamos a **Mediana** (para identificar o comportamento típico) e o **Desvio Padrão** (para medir a variabilidade dos dados). O uso do arredondamento garante uma inspeção visual mais limpa e eficiente.

````python
# Analisando Medianas e Desvio Padrão por Cluster (K-Means 2)

# 1. Agrupar e calcular a mediana
mediana_k2 = df_k2.groupby('cluster_kmeans').median(numeric_only=True).round(2)

# 2. Agrupar e calcular o desvio padrão (apenas colunas numéricas)
desvio_k2 = df_k2.groupby('cluster_kmeans').std(numeric_only=True).round(2)

````

````python
print("--- Medianas por Cluster ---")
display(mediana_k2.T)

print("\n--- Desvio Padrão por Cluster ---")
display(desvio_k2.T)
````

### 📋 Comparação Estrutural: Mediana vs. Desvio Padrão

Para uma análise de perfilamento profunda, é necessário contrastar o **valor típico** (Mediana) com a **volatilidade** (Desvio Padrão) de cada atributo. Utilizamos uma renderização em HTML para exibir essas tabelas lado a lado, facilitando a detecção de padrões e inconsistências.

````python
from IPython.display import display_html

# 1. Definir o estilo: padding lateral (60px) apenas para os valores e cabeçalhos de coluna
estilo_colunas = [
    # th.col_heading foca apenas nos títulos dos clusters (0, 1...)
    dict(selector="th.col_heading", props=[("padding", "10px 60px")]),
    # td foca apenas nas células que contêm os valores numéricos
    dict(selector="td", props=[("padding", "10px 60px")])
]

# 2. Aplicar formatação e o estilo direcionado às colunas
html_mediana = (mediana_k2.T.style
                .format("{:.2f}")
                .set_table_styles(estilo_colunas)
                .set_table_attributes("style='display:inline; margin-right: 80px;'")
                .set_caption("MEDIANAS (K=2)")
                ._repr_html_())

html_desvio = (desvio_k2.T.style
               .format("{:.2f}")
               .set_table_styles(estilo_colunas)
               .set_table_attributes("style='display:inline;'")
               .set_caption("DESVIO PADRÃO (K=2)")
               ._repr_html_())

# 3. Exibir
display_html(html_mediana + html_desvio, raw=True)
````

````python
# Grupos de colunas por escala
cols_monetarias = ['Saldo_Atual', 'Total_Compras', 'Max_Compra_Unica', 'Compras_Parceladas', 'Saque_Adiantado', 'Limite_Credito', 'Total_Pagamentos', 'Pagamento_Minimo']
cols_comportamento = ['Frequencia_Atualizacao_Saldo', 'Frequencia_Compras', 'Frequencia_Compras_Vista', 'Frequencia_Compras_Parceladas', 'Frequencia_Saques_Adiantados', 'Num_Transacoes_Saque', 'Num_Transacoes_Compra', 'Percentual_Pagamento_Integral', 'Meses_Fidelidade']

# Criar Figura
fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(16, 12))

# Subplot 1: Monetários
df_mon_k2 = mediana_k2[cols_monetarias].reset_index().melt(id_vars='cluster_kmeans')
sns.barplot(data=df_mon_k2, x='variable', y='value', hue='cluster_kmeans', palette='coolwarm', ax=ax1)
ax1.set_title('Atributos Monetários - Mediana (K=2)', fontsize=14, fontweight='bold')
ax1.set_ylabel('Valor (R$)')
ax1.tick_params(axis='x', rotation=20)

# Subplot 2: Comportamento
df_comp_k2 = mediana_k2[cols_comportamento].reset_index().melt(id_vars='cluster_kmeans')
sns.barplot(data=df_comp_k2, x='variable', y='value', hue='cluster_kmeans', palette='coolwarm', ax=ax2)
ax2.set_title('Comportamento e Frequência - Mediana (K=2)', fontsize=14, fontweight='bold')
ax2.set_ylabel('Frequência / Contagem')

plt.tight_layout()
plt.show()
````

### 📈 Análise de Tendência Central e Dispersão (Hierárquico K=6)

Diferente do K-Means, o agrupamento hierárquico busca relações de proximidade que podem gerar grupos mais específicos. Nesta etapa, calculamos a **Mediana** e o **Desvio Padrão** para os 6 clusters identificados, permitindo uma análise granular do comportamento dos clientes.

````python
# Analisando Medianas e Desvio Padrão por Cluster (Hierarquico 6)

# 1. Agrupar e calcular a mediana
mediana_h6 = df_h6.groupby('cluster_Hierarquico').median(numeric_only=True).round(2)

# 2. Agrupar e calcular o desvio padrão (apenas colunas numéricas)
desvio_h6 = df_h6.groupby('cluster_Hierarquico').std(numeric_only=True).round(2)
````

### 📋 Comparação Detalhada: Mediana vs. Desvio Padrão (Modelo Hierárquico K=6)

Com o aumento da granularidade para 6 clusters, a exibição lado a lado torna-se ainda mais essencial. Esta visualização permite validar se os novos subgrupos criados pelo algoritmo hierárquico são coesos ou se representam apenas variações ruidosas dos dados.

````python
from IPython.display import display_html

# Estilo ainda mais compacto para comportar 6 colunas
estilo_h6 = [
    dict(selector="th.col_heading", props=[("padding", "8px 18px"), ("font-size", "11px")]),
    dict(selector="td", props=[("padding", "8px 18px"), ("font-size", "11px")]),
    dict(selector="th.row_heading", props=[("padding", "8px 10px"), ("font-size", "11px")])
]

html_mediana_h6 = (mediana_h6.T.style
                   .format("{:.2f}")
                   .set_table_styles(estilo_h6)
                   .set_table_attributes("style='display:inline; margin-right: 20px;'")
                   .set_caption("MEDIANAS (K=6)")
                   ._repr_html_())

html_desvio_h6 = (desvio_h6.T.style
                  .format("{:.2f}")
                  .set_table_styles(estilo_h6)
                  .set_table_attributes("style='display:inline;'")
                  .set_caption("DESVIO PADRÃO (K=6)")
                  ._repr_html_())

display_html(html_mediana_h6 + html_desvio_h6, raw=True)
````
